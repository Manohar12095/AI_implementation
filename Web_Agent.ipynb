```python
from google.colab import userdata
API_KEY = userdata.get('gemini-pro')
print(API_KEY)
```

```python
import google.generativeai as genai
genai.configure(api_key=API_KEY)
print("Gemini API configured successfully.")
```

```python
model = genai.GenerativeModel('gemini-pro')
print("Gemini model initialized successfully.")
```

```python
chat = model.start_chat(history=[])
print("Chat session started.")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Create widgets
input_text = widgets.Textarea(
    value='',
    placeholder='Type your message here...',
    description='Your Message:',
    disabled=False,
    layout=widgets.Layout(height='auto', width='80%')
)
send_button = widgets.Button(
    description='Send',
    disabled=False,
    button_style='',
    tooltip='Send your message',
    icon='send'
)

# Initialize Output widget directly
output_area = widgets.Output()

def on_send_button_clicked(b):
    with output_area:
        clear_output(wait=True) # Clear previous output in the output_area widget
        user_message = input_text.value
        if user_message.strip():
            print(f"You: {user_message}")
            input_text.value = '' # Clear the input after sending

            try:
                response = chat.send_message(user_message)
                print(f"Gemini: {response.text}")
            except Exception as e:
                print(f"Error: {e}")

# Attach the function to the button's click event
send_button.on_click(on_send_button_clicked)

# Display the widgets
display(widgets.VBox([widgets.HBox([input_text, send_button]), output_area]))
print("Interactive chat interface created. Type your message and click 'Send'.")
```
